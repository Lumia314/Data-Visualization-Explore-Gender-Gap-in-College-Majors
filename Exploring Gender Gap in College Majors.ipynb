#Gender gap across college degree
##Data description
The dataset we are working on will have 18 columns with 17 of them being degrees categories and 42 rows showing share of women in those categories from 1970 to 2011. We will visualize our dataset to see the change in female participation in job market between 1970 and 2011.

```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
```

```python
# Overview
women_degrees.info()
women_degrees.describe()
women_degrees[-3:]
```

```python
# Divide the categories into 3 groups of STEM, liberal art and other
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
fig = plt.figure(figsize=(12, 20))
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
# STEM column
for sp in range(0, 18, 3):
    cat_index = int(sp/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[cat_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[cat_index]], c=cb_orange, label='Men', linewidth=3)
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.set_title(stem_cats[cat_index])
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom='off')
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    if cat_index == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif cat_index == 5:
        ax.tick_params(labelbottom='on')
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
# Liberal art
for sp in range(1, 16, 3):
    cat_index = int(sp/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[cat_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[cat_index]], c=cb_orange, label='Men', linewidth=3)
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.set_title(lib_arts_cats[cat_index])
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom='off')
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    if cat_index == 0:
        ax.text(2005, 75, 'Women')
        ax.text(2002, 20, 'Men')
    elif cat_index == 4:
        ax.tick_params(labelbottom='on')
        ax.text(2005, 62, 'Women')
        ax.text(2001, 35, 'Men')
# Other
for sp in range(2, 19, 3):
    cat_index = int(sp/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[cat_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[cat_index]], c=cb_orange, label='Men', linewidth=3)
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.set_title(other_cats[cat_index])
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom='off')
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    if cat_index == 0:
        ax.text(2005, 90, 'Women')
        ax.text(2002, 5, 'Men')
    elif cat_index == 5:
        ax.tick_params(labelbottom='on')
        ax.text(2005, 62, 'Women')
        ax.text(2001, 30, 'Men')
plt.savefig('degrees.png')
```

##Findings:
After we run the above code, we observe a positive change for female roles in most categories:
- Psychology, Physical Science, Biology, Communicaton & Journalism, Social Science, Public Administration, Argriculture, Business, Architecture has seen a clear rise in female participation
- Math & Statistic, Health Profession has seen a more subtle change
- Computer Science, Engineering is still mostly occupied by male

The change can be on account of the feminism movement in 1970s and 1980s. The womens movement was most successful in pushing for gender equality in workplaces and universities.
